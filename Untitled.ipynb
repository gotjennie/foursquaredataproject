```python
import pandas as pd
import numpy as np
```

```python
chi_df = pd.read_csv('./processed_data/chi.csv', index_col=0)
sf_df = pd.read_csv('./processed_data/sf.csv', index_col=0)
nyc_df = pd.read_csv('./processed_data/nyc.csv', index_col=0)
```

```python
all_df = pd.concat([chi_df, sf_df, nyc_df], axis=0)
```

```python
all_df = all_df[all_df['neighborhood'].notnull()]
```

```python
all_df
```

```python
pd.DataFrame({'neighborhood_visits': all_df.groupby(['user_id', 'neighborhood', 'city']).size()}).reset_index()
#checkins_by_user = pd.DataFrame({'count': checkin_df.groupby(['user_id'])['id'].count()}).reset_index()
```

```python
visits_df = pd.DataFrame({'num_visit_type': all_df.groupby(['user_id', 'neighborhood', 'city', 'venue_category']).size()}).reset_index()
```

```python
visits_df
```

```python
categories_df = pd.DataFrame({'num_visit_type': all_df.groupby(['user_id', 'venue_category']).size()}).reset_index()
```

```python
categories_df
```

```python
venue_cateogories_df = pd.DataFrame(
    {'cities': all_df.groupby(['user_id', 'venue_category'])['city'].unique(), 
     'num_cities': all_df.groupby(['user_id', 'venue_category'])['city'].nunique(),
     'neighborhood': all_df.groupby(['user_id', 'venue_category'])['neighborhood'].unique(),
                                    }).reset_index()
```

```python
venue_cateogories_df = venue_cateogories_df[venue_cateogories_df['num_cities'] > 1]
venue_cateogories_df.sort_values('num_cities', ascending=False)
```
